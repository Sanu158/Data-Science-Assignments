# Standardize The Data

```python
nor=StandardScaler()
scaled_data=nor.fit_transform(df1)
scaled_data=pd.DataFrame(scaled_data)
```

# PCA technique to reduce the dimensions of our Data

```python
pca=PCA(n_components=13)
```

```python
model=pca.fit_transform(scaled_data)
```

```python
var=pca.explained_variance_ratio_
var #Variance of each feature in the data
```

As we can see the first 3 columns have high variance when compared to remaining variables, so we can try to consider these columns to build our clustering model.

```python
#Plotting the variances on the graph
plt.scatter(range(1,14), pca.explained_variance_ratio_, color='blue')
plt.plot(range(1,14), pca.explained_variance_ratio_,'g-')
plt.xlabel('Principal Components')
plt.ylabel('Variance %')
plt.xticks(range(1,14))
plt.show()
```

```python
PCA_components=pd.DataFrame(model)
```

```python
plt.scatter(PCA_components[0], PCA_components[1])
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

```python
#Calculating the inertias of clusters
w=[]
for i in range(1,15):
    kmeans=KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(PCA_components.iloc[:,:3])
    w.append(kmeans.inertia_)
```

```python
#Plotting the inertias on the graph
plt.plot(range(1,15), w)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

We can see for 3 clusters have a efficient inertia. So we try to build a clustering model which have 3 clusters

# KMeans Clustering

```python
Kmodel=KMeans(n_clusters=3)
Kmodel.fit(PCA_components.iloc[:,:3])
```

```python
labels=Kmodel.predict(PCA_components.iloc[:,:3])
```

```python
plt.scatter(PCA_components[0], PCA_components[1], c=labels)
plt.show()
```
